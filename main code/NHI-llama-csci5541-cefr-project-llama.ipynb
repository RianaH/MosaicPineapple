```python
# prompt: remove Name: sentence, Length: 100, dtype: object  from each cell

import pandas as pd
import math
import csv
import re

# Replace 'your_file.tsv' with the actual path to your file in Google Drive
file_path = "allData.txt"

df = pd.read_csv(file_path, sep='\t')

# Function to clean the sentence column
def clean_sentence(sentence):
    # Remove "Name:" and everything after it
    sentence = re.sub(r"Name:.*", "", sentence)
    # Remove "Length:" and the number after it
    sentence = re.sub(r"Length: \d+", "", sentence)
    # Remove "dtype: object"
    sentence = re.sub(r"dtype: object", "", sentence)

    #Remove extra whitespace
    sentence = sentence.strip()
    return sentence


df['sentence'] = df['sentence'].apply(clean_sentence)

print(df.head())

def zero_prompt(sentences):
  prompt = f"""
      Classifiy all sentence in the following 100 sentences based on the Common European Framework of Reference for Languages (CEFR) levels: A1, A2, B1, B2, C1, or C2.
      Please evaluate the linguistic complexity, vocabulary, grammar, and overall fluency of each sentence and assign the most appropriate CEFR level.
      Format your output as:
      1. #thesentence: <insert sentence 1 here> #cefrlevel: <insert CEFR level here> #END
      2. #thesentence: <insert sentence 2 here> #cefrlevel: <insert CEFR level here> #END
      ...
      100. #thesentence: <insert sentence 100 here> #cefrlevel: <insert CEFR level here> #END

      Here is the list of sentences:
      {sentences}
      """
  return prompt

num_batch = math.ceil(len(df) / 100)
batches = []
for i in range(num_batch):
  start_index = i * 100
  end_index = min((i + 1) * 100, len(df))
  batches.append(df["sentence"][start_index:end_index])

# Specify the CSV file path
csv_file_path = "prompts.csv"

# Write the prompts to the CSV file
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["prompt"]) #header

    for batch in batches:
        prompt = zero_prompt(list(batch))
        csv_writer.writerow([prompt])

print(f"Prompts have been saved to {csv_file_path}")
```

```python
# prompt: create a list of df from my big df, each has 1000 sentences
# RUN THIS
import pandas as pd
import math

def split_dataframe(filename, chunk_size=1000):
  """Splits a DataFrame into a list of DataFrames with a specified chunk size.

  Args:
    df: The input DataFrame.
    chunk_size: The desired size of each chunk.

  Returns:
    A list of DataFrames.
  """
  df = pd.read_csv(filename, sep='\t')
  chunks = []
  num_chunks = len(df) // chunk_size + (1 if len(df) % chunk_size else 0)  # Calculate number of chunks
  for i in range(num_chunks):
    start = i * chunk_size
    end = min((i + 1) * chunk_size, len(df)) # Ensure end index doesn't exceed the DataFrame length
    chunks.append(df[start:end])
  return chunks


# Assuming your DataFrame is called 'df'
dfs = split_dataframe("allData.txt", 1000)
```
